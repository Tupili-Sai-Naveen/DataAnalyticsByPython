```python
def salary_data(sal):
    if sal=='<=50K':
        return 0
    else:
        return 1
        
```

```python
df['encoded']=df['income'].apply(salary_data)
```

```python
df.head()
```

```python
sns.countplot(x='encoded', data=df)
```

```python
df.replace(to_replace=['<=50K','>50K'],value=[0,1],inplace=True)
```
